```python
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu(x):
    return np.maximum(0, x)

x = np.linspace(-5, 5, 1000) # -5부터 5까지 1000개의 값 생성
y = leaky_relu(x)

print(leaky_relu(-5)) # 0
print(leaky_relu(5)) # 0
print(leaky_relu(100)) # 5

plt.plot(x, y)
plt.title('ReLU Function')
plt.xlabel('Input')
plt.ylabel('Outout')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu(x, alpha=0.01):
    return np.where(x >= 0, x, alpha * x)

x = np.linspace(-5, 5, 100)
y = leaky_relu(x)

print(leaky_relu(-5)) # 0
print(leaky_relu(5)) # 0
print(leaky_relu(100)) # 5

plt.plot(x, y)
plt.title('Leaky ReLU Function')
plt.xlabel('Input')
plt.ylabel('Outout')
plt.show()
```
