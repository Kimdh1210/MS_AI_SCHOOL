```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-10,10,100) # -10부터 10까지 100개의 값 생성
y = sigmoid(x)

print(sigmoid(-10))
print(sigmoid(10))
print(sigmoid(0))
print(sigmoid(100))

plt.plot(x, y)
plt.title('Sigmoid Function')
plt.xlabel('Input')
plt.ylabel('Outout')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def step_fuction(x):
    return  np.array(x>0, dtype=np.int32)

x = np.arange(-5.0, 5.0, 0.1)
y = step_fuction(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.title('Step Function')
```

```python

def derivate_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))

x = np.linspace(-5, 5, 100)
y = derivate_sigmoid(x)
plt.plot(x, y)
plt.show()
```
