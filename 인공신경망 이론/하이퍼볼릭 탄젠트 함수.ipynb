```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

x = np.linspace(-5, 5, 100) # -10부터 10까지 100개의 값 생성
y = tanh(x)

print(tanh(-5))
print(tanh(5))
print(tanh(100))

plt.plot(x, y)
plt.title('Hyperbolic Tangent Function')
plt.xlabel('Input')
plt.ylabel('Outout')
plt.show()
```
