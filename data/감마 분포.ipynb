```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 상황 설정 : 제품 고장까지 걸리는 시간에 관한 분포
# 감마 분포의 주요 파라미터 2개 : 형상 모수 alpha, 비율 모수 beta
alpha = 2.5
beta = 1.2

# 제품의 고장까지 걸리는 시간에 관한 감마분포
lifetime = np.random.gamma(shape=alpha, scale=beta, size=1000)
plt.hist(lifetime, bins=30, density=True, alpha=0.7)
plt.show()
```
