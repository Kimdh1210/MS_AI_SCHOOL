1. 가상 데이터 생성

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

score = []
grades = [1, 2, 3]

for grade in grades:
    for _ in range(100):
        math = np.random.normal(loc=70, scale=6)
        english = np.random.normal(loc=75, scale=4)
        korean = np.random.normal(loc=80, scale=10)
        score.append([grade, math, english, korean])
score_df = pd.DataFrame(score, columns=['Grade', 'Math', 'English', 'Korean'])
score_df
```

2. CSV 파일로 저장

```python
score_df.to_csv('student_scores.csv', index=False)
```

3. 상관 관계 분석 - 히트맵 출력

```python
#csv 파일 불러오기
import pandas as pd
import seaborn as sns
student_scores = pd.read_csv('student_scores.csv')

corr = student_scores.corr()
print(corr)

sns.heatmap(corr, annot=True)
plt.show()
```

4. 과목별 성적 분포와 통계치

```python
print(student_scores['Math'].describe())
print(student_scores['English'].describe())
print(student_scores['Korean'].describe())
```

5. 과목별 성적 분포와 통계치 분석 시각화

```python
plt.figure()
plt.title('Math')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.hist(student_scores['Math'], bins=20)

plt.figure()
plt.title('English')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.hist(student_scores['English'], bins=20)

plt.figure()
plt.title('Korean')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.hist(student_scores['Korean'], bins=20)
```
