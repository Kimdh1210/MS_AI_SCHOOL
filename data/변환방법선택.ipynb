```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

data = pd.DataFrame({'Value':[1, 4, 9, 16, 25]})

plt.hist(data['Value'])
plt.show()
```

분포가 정규분포를 따르지 않는 비정규분포이므로, 비선형 변환을 고려

```python
data['LogValue'] = np.log(data['Value'])
plt.hist(data['LogValue'])
plt.show()
```

```python
transformed_data, Lambda = stats.boxcox(data['Value'])
print(f'최적 람다값 : {Lambda}')
plt.hist(transformed_data)
plt.show()
```
