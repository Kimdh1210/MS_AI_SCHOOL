```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#import os

height = np.random.normal(loc=170, scale=10, size=100)
weight = np.random.normal(loc=70, scale=5, size=100)

data = pd.DataFrame({"height": height, "weight": weight})
#os.makedirs('./data', exist_ok=True)
data.to_csv('human_data.csv', index=False)
```

```python
human_data = pd.read_csv('human_data.csv')

height = human_data['height']
weight = human_data['weight']

plt.hist(height, bins=10, alpha=0.5, label='height')
plt.hist(weight, bins=10, alpha=0.5, label='weight')
plt.show()
```

```python
from scipy import stats

height_mean, height_std = stats.norm.fit(height)
height_x = np.linspace(height.min(), height.max(), 100)
height_y = stats.norm.pdf(height_x, height_mean, height_std)

weight_mean, weight_std = stats.norm.fit(weight)
weight_x = np.linspace(weight.min(), weight.max(), 100)
weight_y = stats.norm.pdf(weight_x, weight_mean, weight_std)

plt.hist(height, bins=10, alpha=0.5, density=True, label='height')
plt.plot(height_x, height_y, label='height distribution')
plt.hist(weight, bins=10, alpha=0.5, density=True, label='weight')
plt.plot(weight_x, weight_y, label='weight distribution')
plt.legend(loc='upper right')
plt.show()
```
