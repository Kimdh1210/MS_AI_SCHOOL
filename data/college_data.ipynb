```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('college_data.csv')
data
```

```python
print(data['top10perc'].unique())
```

```python
plt.figure()
plt.hist(data['top10perc'], bins=20)
plt.title("Histogram of Top 10% Scores")
plt.xlabel("Top 10% Scores")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
data.groupby("private")['top10perc'].sum().plot(kind='pie')
plt.title("Pie Chart of Top 10% Scores by Private/Public")
plt.show()
```
