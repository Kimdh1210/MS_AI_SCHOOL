```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# F-분포의 주요 파라미터 : DOF1, DOF2
# DOF1 : 분자의 자유도. 검정에 참여하는 그룹의 자유도
# DOF2 : 검정을 진행하는 샘플 그룹의 표본 갯수 - 분자의 자유도
dof1 = 5
dof2 = 10

f_values = np.random.f(dof1, dof2, 1000)

plt.hist(f_values, bins=20, alpha=0.5)
plt.show()
```
