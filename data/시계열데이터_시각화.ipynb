```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dates = pd.date_range(start='2023-01-01', periods=1000)
random_values = np.random.randn(1000)
cumsum_values = np.random.randn(1000).cumsum()

data = pd.DataFrame({
    'Dates': dates,
    'Random_Values': random_values,
    'Cumsum_Values': cumsum_values
})

plt.plot(data['Dates'], data['Random_Values'])
plt.show()
plt.figure()
plt.plot(data['Dates'], data['Cumsum_Values'])
plt.show()
```

```python
plt.figure()
plt.scatter(data['Dates'], data['Random_Values'])
plt.show()
plt.figure()
plt.scatter(data['Dates'], data['Cumsum_Values'])
plt.show()
```

```python
plt.bar(data['Dates'], data['Random_Values'])
plt.show()
plt.bar(data['Dates'], data['Cumsum_Values'])
plt.show()
```

```python
plt.boxplot(data['Random_Values'])
plt.show()
plt.boxplot(data['Cumsum_Values'])
plt.show()
```

```python
plt.hist(data['Random_Values'])
plt.show()
plt.hist(data['Cumsum_Values'])
plt.show()
```

```python
import seaborn as sns

new_dates = pd.date_range(start='2023-01-01', periods=100)
new_random_values = np.random.randn(100)
new_cumsum_values = np.random.randn(100).cumsum()

new_data = pd.DataFrame({
    'Dates': new_dates,
    'Random_Values': new_random_values,
    'Cumsum_Values': new_cumsum_values
})

pivot_table_random = new_data.pivot(index='Dates', columns='Random_Values', values='Random_Values')
pivot_table_cumsum = new_data.pivot(index='Dates', columns='Cumsum_Values', values='Cumsum_Values')

sns.heatmap(pivot_table_random, cmap='YlOrRd', cbar=True)
plt.show()
sns.heatmap(pivot_table_cumsum, cmap='YlOrRd', cbar=True)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result_random = seasonal_decompose(data['Random_Values'], period=365)
result_cumsum = seasonal_decompose(data['Cumsum_Values'], period=365)

plt.plot(result_random.trend)
plt.title('Random Data Trend')
plt.show()
plt.plot(result_cumsum.trend)
plt.title('Cumsum Data Trend')
plt.show()

plt.plot(result_random.seasonal)
plt.title('Random Data Seasonality')
plt.show()
plt.plot(result_cumsum.trend)
plt.title('Cumsum Data Seasonality')
plt.show()

plt.plot(result_random.resid)
plt.title('Random Data Residuals')
plt.show()
plt.plot(result_cumsum.resid)
plt.title('Cumsum Data Residuals')
plt.show()
```
