```python
import numpy as np
import matplotlib.pyplot as plt

# 상황 : 고객 서비스 센터 대기 시간 분포 측정

# 지수분포에 필요한 파라미터 lambda : 다음으로 사건이 발생할 때까지 걸릴 평균 시간
lambda_scale = 0.5

#고객 서비스 센터 대기 시간을 모사하는 지수분포
waiting_times = np.random.exponential(scale=1/lambda_scale, size=1000)

plt.hist(waiting_times, bins=30, density=True, alpha=0.7)
plt.show()
```
