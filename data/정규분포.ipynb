```python
import numpy as np
import matplotlib.pyplot as plt

# 정규 분포를 따르는 난수 데이터 생성
# 평균은 0이고 분산은 1인 정규분포
data = np.random.normal(loc=0, scale=1, size=10000)

plt.hist(data, bins=100)
plt.show()
```

```python
# 위에서 생성한 난수행렬의 평균값
mean = np.mean(data)
# 난수행렬의 표준편차
std = np.std(data)
print(f"평균 : {mean}")
print(f"표준편차 : {std}")
```
