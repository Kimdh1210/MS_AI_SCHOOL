```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dates = pd.date_range('20230101', periods=100)
values = np.random.randn(100) * 10

data = pd.DataFrame({'Date': dates, 'Value': values})

plt.plot(data['Date'], data['Value'])
plt.show()
```

```python
WINDOW_SIZE = 7
data['MovingAverage'] = data['Value'].rolling(window=WINDOW_SIZE).mean()
plt.plot(data['Date'],data['Value'])
plt.plot(data['Date'],data['MovingAverage'])
plt.show()
```

```python
# 지수평활법의 평활 계수
alpha = 0.3
# 지수평활법 계산
data['Exponential_Smoothing'] = data['Value'].ewm(alpha=alpha).mean()

plt.plot(data['Date'], data['Value'], label='Original')
plt.plot(data['Date'], data['Exponential_Smoothing'], label='Exp_Smoothing', color='red')
plt.plot(data['Date'], data['MovingAverage'], label='MovingAverage', color='blue')
plt.legend()
plt.show()
```
