```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 정규분포 행렬에 이상치에 해당하는 작은 행렬을 앞뒤로 덧붙임
data = np.concatenate([np.random.normal(0,1,900),
                       np.random.normal(8,2,100),
                       np.random.normal(-4,2,100)])
plt.hist(data, bins=50)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
plt.boxplot(data)
plt.xlabel("Data")
plt.ylabel("Values")
plt.show()
```
