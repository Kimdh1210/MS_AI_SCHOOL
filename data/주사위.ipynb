```python
import numpy as np
import matplotlib.pyplot as plt

# 주사위 던지기를 모의하는 코드이므로, 1~6 범위가 필요
# randint 함수의 high 인자는 실제 최대값 + 1을 받아야함
data = np.random.randint(1, 6+1, size=100)
print(f"max of data : {data.max()}")
```

```python
# 빈도수 계산 (counts에 빈도가 들어감)
unique, counts = np.unique(data, return_counts=True)

plt.bar(unique, counts)
plt.xlabel("Dice Number")
plt.ylabel("Frequency")
plt.show()
```

```python
value_counts = np.bincount(data)
print(value_counts)
```

```python
relative_frequencies = value_counts / len(data)
print(relative_frequencies)
```

```python
plt.figure()
plt.bar(np.arange(6), relative_frequencies[1:])
plt.xlabel("Dice number")
plt.ylabel("Relative frequency")
plt.show()
```

```python
plt.figure()
plt.pie(relative_frequencies[1:], labels=np.arange(7)[1:],
        autopct="%1.1f.%%")
plt.show()
```
