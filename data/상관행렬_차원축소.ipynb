```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
z = np.random.normal(0, 1, 100)

data = np.vstack([x, y, z]).T

corr_mat = np.corrcoef(data, rowvar=False)
```

```python
sns.heatmap(corr_mat, annot=True, cmap='coolwarm')
plt.show()
```

```python
# 다중공선성 판단 기준점
threshold = 0.1

selected_vars = []
for i in range(corr_mat.shape[0]):
    for j in range(i + 1, corr_mat.shape[1]):
        if abs(corr_mat[i, j]) >= threshold:
            selected_vars.extend([i, j])
        
if selected_vars: # selected_vars 내용물이 존재하는 경우
    reduced_data = data[:, list(set(selected_vars))]
    plt.scatter(reduced_data[:, 0], reduced_data[:, 1])
    plt.show()
else : # selected_vars가 빈 list인 경우
    print('다중공선성이 존재하지 않는 것으로 판단됨')
```
