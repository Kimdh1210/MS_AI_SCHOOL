```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 정규분포 데이터는 중앙값이 평균과 거의 동일하기 때문에,
# 구분될 수 있는 지수분포 데이터를 생성
data = np.random.exponential(size=10000)

mean = np.mean(data)
median = np.median(data)
# 최빈값은 numpy에서 데이터에 대해서 빠르게 얻을 수 있는 편의함수가 없음
# 보통 scipy 모듈에 포함된 stats.mode 함수를 사용
mode = stats.mode(data)

print(f"평균 : {mean}")
print(f"중앙값 : {median}")
print(f"최빈값 : {mode}")
```

```python
# numpy로 만든 난수행렬은 연속형 데이터로, 
# 중복값이 나오지 않기 때문에 최빈값이 의미가 없음
```

```python
plt.hist(data, bins=50)
# 평균에 해당하는 붉은색 점선
plt.axvline(mean, color='red', linestyle="dashed", linewidth=2)
# 중앙값에 해당하는 녹색 점선
plt.axvline(median, color='green', linestyle="dashed", linewidth=2)

plt.legend([f"Mean : {mean:.2f}", f"Median: {median:.2f}"])

plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Exponential Data")
plt.show()
```

```python
from scipy import stats
plt.figure()
# 연속형 데이터가 아닌, 이산 데이터에 해당하는 정수 집합
data_int = np.random.normal(loc=0, scale=100, size=10000)
# numpy 행렬로 얻는 연속형 데이터 변수 data_int
# data_int에는 round() 함수가 내장되어 있으며, 해당 함수를 사요하면
# 내부값들에 대해 반올림 연산을 진행하여 모두 정수로 바꿈
data_int = data_int.round()
mode = stats.mode(data_int)
print(f"최빈값 : {mode}")

plt.hist(data_int, bins=50)
# stats.mode로 얻은 결과물인 위의 변수 mode는
# 정수값이 아니라 최빈값과 개수를 포함하는 ModeResult값임
# 따라서 x축에 지정할 좌표인 최빈값 자체를 얻기 위해
# 해당 변수가 가진 하위 속성값 mode를 다시 불러와야 함
plt.axvline(mode.mode, color="purple", linestyle="dashed",
            linewidth=2)
plt.xlabel("Data Values")
plt.ylabel("Frequency")
plt.show()
```
