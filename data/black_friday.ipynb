```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#데이터셋 불러오기
data = pd.read_csv('BlackFriday.csv')
```

```python
#User_ID별 총 구매 금액 계산
user_total_spent = data.groupby('User_ID')['Purchase'].sum()
```

```python
# 50 단위로 묶은 히스토그램 그리기
plt.hist(user_total_spent, bins=50)
plt.title("Histogram of Total Purchase Amount by User")
plt.xlabel("Total Purchase Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
# Pandas 자료형 자신의 밀도 그래프 작성
user_total_spent.plot.kde()
plt.title("Density Plot of Purchase Amount by User")
plt.xlabel("Total Purchase Amount")
# 밀도 그래프의 y축은 밀도이며, 0~1 사이의 값을 갖는 그래프가 출력됨
plt.show()
```
