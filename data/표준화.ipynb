```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from data_utils import create_date_df

values = [10, 12, 11, 15, 9, 13, 8, 14, 10, 16] * 10

data = create_date_df(100, values)
data
```

```python
scaler = StandardScaler()

# StandardScaler에서 요구하는 칼럼값의 형태가 (n, 1) 형태의 2차원 numpy array이므로
# Value 칼럼의 내용물을 받아서 reshape(-1 ,1)을 사용하여 알맞는 형태로 변환
to_standardize = data['Value'].values.reshape(-1,1)

data['Value_Standardized'] = scaler.fit_transform(to_standardize)

import matplotlib.pyplot as plt

plt.plot(data['Date'], data['Value_Standardized'])
plt.show()
```

```python
data['Value_Scaled'] = (data['Value'] - data['Value'].mean()) / data['Value'].std()
print(data['Value_Scaled'], data['Value_Standardized'])
```
