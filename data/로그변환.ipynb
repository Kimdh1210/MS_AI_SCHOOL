```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.DataFrame({'Value' : [10, 100, 1000, 10000, 100000]})

# 로그 변환
data['LogValue'] = np.log(data['Value'])
print(data['LogValue'])
```

```python
plt.plot( data['Value'], color='blue')
plt.show()
plt.plot(data['LogValue'], color='orange')
plt.show()
```
