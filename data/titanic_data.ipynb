```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic_df = pd.read_csv('Titanic_data.csv')
```

```python
# 성별에 따른 생존률
survived_by_sex =titanic_df.groupby('Sex')['Survived'].mean()

sns.barplot(x=survived_by_sex.index, y=survived_by_sex.values)
plt.title("Survival Rate by gender")
plt.xlabel("gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
# 탑승 항구에 따른 생존자 수
survived_embarked = titanic_df.groupby('Embarked')['Survived'].sum()
# 탑승 항구로 그룹지었을 때, "생존" 카테고리의 총 길이
# 즉 해당 항구의 총 탑승객 수에서 생존자를 감산하면 사망자 수
dead_embarked = titanic_df.groupby('Embarked')['Survived'].count() - survived_embarked

plt.figure()

plt.bar(survived_embarked.index, survived_embarked.values,
        color='g', alpha=0.5)
plt.bar(dead_embarked.index, dead_embarked.values,
        bottom=survived_embarked.values, color='r', alpha=0.5)
plt.xlabel("Embarked Port")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
# 객실 등급에 따른 생존자 수
survived_by_class = titanic_df.groupby("Pclass")["Survived"].sum()
dead_by_class = titanic_df.groupby("Pclass")["Survived"].count() - survived_by_class

plt.figure()
plt.bar(survived_by_class.index, survived_by_class.values,
        color='g', alpha=0.5)
plt.bar(dead_by_class.index, dead_by_class.values,
        bottom=survived_by_class, color='r', alpha=0.5)
plt.show()
```
