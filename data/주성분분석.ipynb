```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

data = np.random.rand(100, 3) # 각 샘플 100개로 구성된 변수 3개가 포함된 데이터

# PCA 모델 생성 및 적합
pca = PCA(n_components=2) # 주성분 2개에 대해 분석 진행
pca.fit(data)

projected_data = pca.transform(data) # 차원 축소 진행

print(f'주성분의 설명력 : {pca.explained_variance_}') # 고유벡터의 고유값
print(f'주성분의 설명력 비율 : {pca.explained_variance_ratio_ * 100} %') # 주성분이 데이터 분산을 얼마나 설명하는가?

print(f'주성분의 가중치 : {pca.components_}')
```

```python
plt.scatter(projected_data[:,0], projected_data[:,1])
plt.show()
```
