```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 베타 분포의 파라미터 : alpha, beta
alpha = 2
beta_param = 5

# 무작위 사용자 행동 데이터 생성 (베타 함수 정의상 0~1 사이의 값)
user_data = np.random.beta(alpha, beta_param, size=1000)

plt.hist(user_data, bins=30, density=True, alpha=0.7)
plt.show()
```
