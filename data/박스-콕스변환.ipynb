```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

data = pd.DataFrame({'Value': [10, 20, 30, 40, 50]})

# lambda는 python에 내장된 키워드이기 때문에 (람다 함수) 변수명으로 사용할 수 없음에 주의
transformed_data, Lambda = stats.boxcox(data['Value'])

print(transformed_data)
print(f'최적 람다값 : {Lambda}')
```

```python
plt.plot(transformed_data, label='Transformed data')
plt.plot(data['Value'], label='Original data')
plt.legend()
plt.show()
```
