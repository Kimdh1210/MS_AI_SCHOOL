```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

size = 100
data = np.random.normal(0, 1, size)

plt.plot(data)
plt.show()
```

```python
best_aic = np.inf
best_order = None

# 무차별 대입으로 모든 차수에 대해 aic 확인
# (2, 1, 2 범위 안에서 탐색)
# ARIMA(p, d, q)
for p in range(3):
    for d in range(2):
        for q in range(3):
            order = (p, d, q)
            model = ARIMA(data, order=order)
            results = model.fit()
            if results.aic < best_aic:
                best_aic = results.aic
                best_order = order

print(f'Best AIC : {best_aic}')
print(f'Best Order : {best_order}')
```
