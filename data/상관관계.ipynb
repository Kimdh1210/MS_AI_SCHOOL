```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(loc=0, scale=1, size=100)
y = (2 * x) + np.random.normal(loc=0, scale=1, size=100)
# y = 2x + a

plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#상관계수 계산
corr = np.corrcoef(x, y)
print("상관계수 : ")
print(corr)
#x, y에 대한 상관관계를 출력하면 다음과 같이 나온다.
# [[1.         0.90906692]
# [0.90906692 1.        ]]
# 대각선 성분은 각자 자기 자신과의 상관관계이므로 무조건 1이 나오며,
# 실제로 두 변수 사이에 발생하는 상관계수는 나머지 값이 된다.
print(f"상관계수 : {corr[0][1]}")
```
