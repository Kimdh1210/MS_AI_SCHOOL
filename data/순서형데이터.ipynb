```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('college_data.csv')
```

```python
# data의 admission_level 컬럼을 새로 선언
# top10perc 컬럼을 기준으로 4개 그룹으로 나뉜 레벨을 포함한 컬럼
data['admission_level'] = pd.qcut(data['top10perc'], q=4, labels=['very_low', 'low', 'high', 'very_high'])
print(data)
```

```python
# 위에서 생성한 admission_level에 대해 서로간의 빈도수 대소를 파악하기 위한 막대 그래프
plt.bar(data['admission_level'].value_counts().index, data['admission_level'].value_counts().values)
plt.xlabel('Admission level')
plt.ylabel('Number of Colleges')
plt.show()
```
