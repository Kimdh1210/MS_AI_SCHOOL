```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
# 입력 데이터 생성
input_data = torch.randn(1, 1, 28, 28) # 배치 크기 1, 채널 1, 이미지 크기 28x28

# 스트라이드 컨볼루션을 적용한 컨볼루션 레이어 정의
conv = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3, stride=2, padding=1)

# 컨볼루션 연산
output = conv(input_data)

# 결과 출력
print(output.shape)
```

```python
# 결과 시각화
plt.subplot(1,2,1)
plt.imshow(input_data.squeeze(), cmap='gray')
plt.title("Input")

plt.subplot(1,2,2)
plt.imshow(output.squeeze().detach().numpy()[0], cmap='gray')
plt.title("Output")

plt.tight_layout()
plt.show()
```
