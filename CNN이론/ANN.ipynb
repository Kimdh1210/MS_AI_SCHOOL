```python
import torch
import torch.nn as nn
```

```python
class ANN(nn.Module) : 
    def __init__(self, input_size, hidden_size, output_size) : 
        super(ANN, self).__init__()
        """
        ANN 구조 : input -> hidden + 활성함수 -> output 
        """ 
        
        self.fc1 = nn.Linear(input_size, hidden_size) # 입력층 - 은닉층 (선형 변환을 적용)
        self.relu = nn.ReLU()                         # 활성함수 추가 
        self.fc2 = nn.Linear(hidden_size, output_size)# 은닉층 - 출력층 (선형 변환을 적용)
        
    
    def forward(self, x) :  # 입력 데이터 X 를 받아서 순전파 연산을 수행 하는 곳 
        out = self.fc1(x) # 입력층 -> 은닉층 연산 적용 
        out = self.relu(out) # 은닉층에서 나온 값을 활성화 함수 적용 
        out = self.fc2(out) # 은닉층 -> 출력층 연산 수행 
        
        return out
```

```python
# 모델 인스턴스 생성 
input_size = 784
hidden_size = 256
output_size = 10

model = ANN(input_size, hidden_size, output_size)

# 손실 함수 정의 
criterion = nn.CrossEntropyLoss()

# 최적화 알고리즘 정의 
learning_rate = 0.01 
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
# 학습 데이터와 레이블 생성 
inputs = torch.randn(100, input_size) # 입력 데이터 (크기 100 X 784)
labels = torch.randint(0, output_size, (100,)) # 레이블 (크기 : 100)
```

```python
num_epochs = 100
for epoch in range(num_epochs) : 
    
    # Forward
    outputs = model(inputs)
    loss = criterion(outputs, labels)
    
    # Backward 
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    # 로그 출력 
    if (epoch+1) % 1 == 0 : 
        print(f"Epoch [{epoch+1}/{num_epochs}] Loss: {loss.item():.4f}")
```
