## Series
* 1차원 배열과 같은구조
* 값을 저장하는 value와 색인를 저장하는 index로 구성
* 동일한 종류의 자료를 저장

```python
import pandas as pd
import numpy as np
```

```python
s = pd.Series(np.random.randn(5))
print(s)
```

```python
s1 = pd.Series(np.random.randn(5), index=(range(1,6)))
print(s1)
```

```python
s2 = pd.Series(np.random.randn(7), index = ['s', 'm', 't', 'w', 't','f','s'])
print(s2)
```

```python
df2 = pd.DataFrame({'A': 1.,
                    'B': pd.Timestamp('20130102'),
                    'C': pd.Series(1, index=list(range(4)), dtype='float32'),
                    'D': np.array([3]*4, dtype='int32'),
                    'E':['test', 'train', 'test', 'train'],
                    'F': 'foo'})
print(df2)
```

### 데이터 합치기
 
 

#### concat

```python
df = pd.DataFrame(np.random.randn(10,4), columns=list('ABCD'))
print(df)
```

```python
pieces = [df[:3], df[3:7], df[7:]]
print(pieces[0], end='\n==========================================\n')
print(pieces[1], end='\n==========================================\n')
print(pieces[2], end='\n==========================================\n')
```

```python
pd.concat(pieces)
```

#### join

```python
left = pd.DataFrame({'key':['A','B'], 'lval':[1,2]})
print(left)
```

```python
right = pd.DataFrame({'key':['A','B'], 'rval':[4,8]})
print(right)
```

```python
merged = pd.merge(left, right)
print(merged)
```

```python
setosa = iris[iris.Species=='setosa']
versicolor = iris[iris.Species == 'versicolor']
virginica = iris[iris.Species == 'virginica']
```

```python
setosa1 = setosa[['PetalLength','PetalWidth']]
print(setosa1)
versicolor1 = versicolor[['PetalLength','PetalWidth']]
print(versicolor1)
virginica1 = virginica[['PetalLength','PetalWidth']]
print(virginica1)
```

```python
setosa_n = setosa1.to_numpy()
print(setosa_n[0:3], setosa_n.shape)
versicolor_n = versicolor1.to_numpy()
print(versicolor_n[0:3], versicolor_n.shape)
virginica_n = virginica1.to_numpy()
print(virginica_n[0:3], virginica_n.shape)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(3,3))
plt.title('Scatter Plot')
plt.scatter(setosa_n[:,0], setosa_n[:,1])
plt.scatter(versicolor_n[:,0], versicolor_n[:,1])
plt.scatter(virginica_n[:,0], virginica_n[:,1])
plt.legend(['setosa','versicolor','virginica'])
plt.xlabel('PetalLength')
plt.ylabel('PetalWidth')
plt.show()
```
