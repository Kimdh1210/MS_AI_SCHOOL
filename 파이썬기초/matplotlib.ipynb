# Matplotlib
* 파이썬용 시각화 라이브러리 이다.
* 라인플롯, 바 차트, 파이차트, 히스토그램 등의 차트와 플롯 등 다양한 스타일 지원
* 머신러닝시 데이터 분석이나 결과 확인을 위한 시각화에 사용

### matplotlib를 import 해서 간단한 Line Chart 그리기

* 코드를 간결하게 사용하기 위해서 matplotlib의 pyplot을 import하고 plt라고 alias를 설정

* plt.plot([x데이터], y데이터)

* '[ ]'(대괄호)안에 있는 내용은 생략가능, x데이터가 생략될 경우에 x데이터는 0부터 y데이터의 갯수만큼 1씩증가




```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.plot([4,6,7,9])
plt.show()
```

### 차트 제목, 축 이름 설정

```python
x = ['Seoul','Tokyo','London','NewYork']
y = [4,5,6,7]
plt.plot(x,y)
plt.xlabel('city')
plt.ylabel('income')
plt.title('result')
plt.show()
```

### 복수의 데이터사용

## figure 
    *모든 plot요소의 최상위 콘테이너

### 범례(legend) 설정

```python
plt.figure(figsize=(3,2))
plt.plot([1,2,3],[4,5,6])
plt.plot([1,3,5],[2,4,6])
plt.plot([2,3,6],[1,3,5])
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.legend(['a','b','c'])
plt.show()
```

### plot함수의 매개변수들

* plot(\*args, scalex=True, scaley=True, data=None, \**kwargs)
* 매개변수에 \* 하나인것은 tuple
* 매개변수에 \** 두개인것은 dictionary

#### args

plot([x데이터1], y데이터1, [fmt], [x데이터2], y데이터2, [fmt], [x데이터3], y데이터3, [fmt],.....)

```python
plt.figure(figsize=(3,3))
plt.plot([1,2,3],[3,4,5],'g--v',[2,3,4],[1,2,3])
plt.show()
```

#### data - 딕셔너리로 구성된 데이터를 입력할때 사용


```python
dic_data = {'x_data':[1,2,3,4], 'y_data':[4,5,6,7]}
plt.figure(figsize=(3,3))
plt.plot('x_data', 'y_data', data = dic_data)
plt.show()
```

### Pie chart

* ax.pie()
- x: 데이터
- labels: 레이블 배열
- colors: 색상 배열

### scatter (산포도)

* ax.scatter()
- x: x데이터
- y: y데이터
- c: marker 색상
- marker: 마커형태 (. o * + D)
- alpha: 마커투명도(0~1)
- linewidths: 마커 테두리 굵기
- edgecolors: 마커 테두리 색깔

## histogram

* ax.hist(x, bin=10, color="#FF0000", ec="#FFFFFF", alpah=0.5)
- x: 데이터배열
- bins: 계급 수, 'auto'
- color: 색상 16진수 ex)"#FF0000"
- ec: 경계선의 색상 16진수
- alpha: 투명도(0~1)
