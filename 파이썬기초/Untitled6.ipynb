```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt, image as img
```

```python
image = Image.open('logo.jpg')
print(image.format)
print(image.mode)
print(image.size)
```

```python
image.show()
plt.imshow(image)
plt.show()
```

```python
new_image = image.resize((600,600))
new_image.show()
new_image.save('logo600.jpg')
plt.imshow(new_image)
plt.show()
```

```python
new_image1 = image.transpose(Image.FLIP_LEFT_RIGHT)
new_image1.show()
new_image1.save('logo_flit1.jpg')
plt.imshow(new_image1)
plt.show()
```

```python
new_image1 = image.transpose(Image.FLIP_TOP_BOTTOM)
new_image1.show()
new_image1.save('logo_tlip2.jpg')
plt.imshow(new_image1)
plt.show()
```

```python
gray_image = image.convert("L")
gray_image.save('gray_logo.jpg')
gray_image.show()
```

```python
image1 = img.imread('logo.jpg')
plt.imshow(image1)
plt.show()
```
