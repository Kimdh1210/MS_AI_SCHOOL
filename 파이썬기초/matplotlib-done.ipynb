### matplotlib를 import 해서 간단한 Line Chart 그리기

* 코드를 간결하게 사용하기 위해서 matplotlib의 pyplot을 import하고 plt라고 alias를 설정

* plt.plot([x데이터], y데이터)

* '[ ]'(대괄호)안에 있는 내용은 생략가능, x데이터가 생략될 경우에 x데이터는 0부터 y데이터의 갯수만큼 1씩증가

```python
plt.plot([4, 6, 7, 9]) 
plt.show()
```

### subplot
* 두개이상의 차트가 필요할때 사용

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

fig = plt.figure(figsize=(5,4))
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```

```python
x = np.arange(1,10)
y1 = x*5
y2 = x*1
y3 = x*0.3
y4 = x*0.2

plt.figure(figsize=(5,3))
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.show()
```

```python
fig = plt.figure(figsize=(2,2))
ax1 = fig.add_subplot(1,1,1)
x = list(range(5))
y = list(range(5))
ax1.plot(x, y)
plt.show()
```

```python
datas = {'x_value':[1,2,3,4], 'value1':[1,5,6,7],'value2':[2,6,5,3],'value3':[3,9,6,7],'value4':[4,5,6,7]}

fig = plt.figure(figsize=(5,3))

ax1 = fig.add_subplot(2,2,1)
ax1.set_title('#1')
ax1.plot(datas['x_value'],datas['value1'])

ax2 = fig.add_subplot(2,2,2)
ax2.set_title('#2')
ax2.plot(datas['x_value'],datas['value2'])

ax3 = fig.add_subplot(2,2,3)
ax3.set_title('#3')
ax3.plot(datas['x_value'],datas['value3'])

ax4 = fig.add_subplot(2,2,4)
ax4.set_title('#4')
ax4.plot(datas['x_value'],datas['value4'])
plt.show()
```

```python
datas = {'x_value':[1,2,3,4], 'value1':[1,5,6,7],'value2':[2,6,5,3],'value3':[3,9,6,7],'value4':[4,5,6,7]}

fig, axes = plt.subplots(2,2,figsize=(5,3))
axes[0][0].plot(datas['x_value'],datas['value1'])
axes[0][1].plot(datas['x_value'],datas['value2'])
axes[1, 0].plot(datas['x_value'],datas['value3'])
axes[1, 1].plot(datas['x_value'],datas['value4'])
plt.show()
```

### Bar chart

```python
y = [5, 3, 7, 10, 9, 5, 3.5, 8]
x = range(len(y))
plt.figure(figsize=(4, 4))
plt.bar(x, y, width=0.7, color="blue")
plt.show()
```

### Pie chart

* ax.pie()
- x: 데이터
- labels: 레이블 배열
- colors: 색상 배열

```python
labels = ['Apple', 'Samsumg', 'LG', 'HUAWEI']
sizes = [35, 25, 10, 30]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0.2, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.2f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

### scatter (산포도)

* ax.scatter()
- x: x데이터
- y: y데이터
- c: marker 색상
- marker: 마커형태 (. o * + D)
- alpha: 마커투명도(0~1)
- linewidths: 마커 테두리 굵기
- edgecolors: 마커 테두리 색깔

```python
x=np.random.normal(0, 1, 100)
y=np.random.normal(0, 1, 100)

plt.figure(figsize=(3,3))
plt.title('Scatter Plot')
plt.scatter(x, y)
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(4,4))
axes[0, 0].scatter(x,y, marker='o')
axes[0, 1].scatter(x,y, s=20, marker='*', edgecolors='red', c='blue')
axes[1, 0].scatter(x,y, alpha=0.5, marker='+')
axes[1, 1].scatter(x,y, marker='.', edgecolor='orange')
plt.show()
```

## histogram

* ax.hist(x, bin=10, color="#FF0000", ec="#FFFFFF", alpah=0.5)
- x: 데이터배열
- bins: 계급 수, 'auto'
- color: 색상 16진수 ex)"#FF0000"
- ec: 경계선의 색상 16진수
- alpha: 투명도(0~1)


```python
x = np.random.normal(0, 10, 1000)
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(1,1,1)
ax.hist(x, bins='auto', ec='#FFFFFF' )
plt.show()
```
