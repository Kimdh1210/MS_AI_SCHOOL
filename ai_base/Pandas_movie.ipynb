```python
import numpy as np
import pandas as pd


def get_ratings_numpy(df):
    # ============================================
    # TODO : datafame  , numpy    () .
    #      [input] : dataframe of pandas 
    #       
    #      [output] :  id    data
    arr = np.array(df)      # Hint
    # ================ EDIT HERE =================
    ratings = {}
    for i in np.unique(arr[:,1]):
        tmp = arr[np.where(arr[:,1] == i )]
        ratings[i] = np.mean(tmp[:,2])
    
    # ============================================
    return ratings

def get_ratings_pandas(df):
    # ============================================
    # TODO : pandas  get_ratings_numpy     .
    #        ,    .
    #      [input] : dataframe of pandas 
    #       
    #      [output] :  id    data
    # ================ EDIT HERE =================
    id_groups = df.groupby('movieId').mean()
    ratings = id_groups['rating']
    # ============================================
    return ratings


if __name__ == '__main__':
    ratings_path = 'ratings.csv'
    #movies_path = '\\movies.csv'
    ratings = pd.read_csv(ratings_path)
    #movies = pd.read_csv(movies_path)

    # ===============    ================
    ratings_n = get_ratings_numpy(ratings)
    ratings_p = get_ratings_pandas(ratings)
    # ============================================
    

    print(ratings_n )
    print(ratings_p)
```
